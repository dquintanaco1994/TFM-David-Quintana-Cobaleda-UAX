```python
# Preparamos el entorno para poder entrenar el modelo, procesar las imágenes y mostrar los resultados directamente en el cuaderno
from ultralytics import YOLO
import os
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import time
import numpy as np
```

```python
# Localizamos la carpeta de trabajo, establecemos la ruta al archivo de datos y definimos dónde se guardarán los resultados del entrenamiento
BASE_DIR   = os.getcwd()
DATA_YAML  = os.path.join(BASE_DIR, "data.yaml")
SAVE_DIR   = os.path.join(BASE_DIR, "runs", "train")
```

```python
# Iniciamos el entrenamiento de un modelo preentrenado durante 50 ciclos con lotes de 16 imágenes a 640×640, usando GPU y guardando los resultados en la carpeta indicada
model = YOLO('yolo11s.pt')
model.train(
    data      = DATA_YAML,
    imgsz     = 640,
    batch     = 16,
    epochs    = 50,
    device    = "0",
    project   = SAVE_DIR,
    name      = "yolo11_exp",
    exist_ok  = True
)
```

```python
model = YOLO(r"runs/train/yolo11_exp/weights/best.pt")
```

```python
# Evaluamos el modelo entrenado usando el conjunto de prueba para obtener métricas de rendimiento
results = model.val(
    data="data.yaml",
    batch=16,
    imgsz=640,
    device="0",
    split="test" 
)
```

```python
# Esta función carga una imagen, aplica el modelo de detección para identificar objetos y muestra el resultado con las anotaciones
def detect_and_show(model, img_path: str):
    # Lee la imagen
    img = cv2.imread(img_path)
    if img is None:
        raise FileNotFoundError(f"No encuentro la imagen: {img_path}")

    # Inferencia y dibujo 
    results = model(img)[0]
    annotated = results.plot()

    # Convierte a RGB y muestra
    annotated_rgb = cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8,8))
    plt.imshow(annotated_rgb)
    plt.axis('off')
    plt.show()
```

```python
# Función para medir el rendimiento de inferencia de un modelo en imágenes por segundo (FPS)
def benchmark(model, img: np.ndarray, runs: int = 50):
    for _ in range(5):
        _ = model.predict(img)
    start = time.time()
    for _ in range(runs):
        _ = model.predict(img)
    total = time.time() - start
    fps = runs / total
    print(f"Procesamiento medio: {fps:.2f} FPS")
    return fps
```

```python
detect_and_show(model, r"Aviones.jpg")
```

```python
benchmark(model, r"Aviones.jpg")
```

```python
detect_and_show(model, r"Pajaros.jpg")
```

```python
benchmark(model, r"Pajaros.jpg")
```
